```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
data = pd.read_excel(r'SPX Data.xlsx')   #importing data
data = pd.DataFrame(data)
```

```python
#visualzing Our Dataset
plt.figure(figsize = (12.5,4.5))
plt.plot(data['Close Price'])
plt.xlabel('1980/01/02 - 2019/11/25')
plt.ylabel('closing price')
plt.grid()
plt.show()
```

```python
# Creating Simple Moving Averages
SMA = pd.DataFrame() 
SMA['SMA30'] = data['Close Price'].rolling(window=30).mean()       
SMA['SMA300'] = data['Close Price'].rolling(window=300).mean()
```

```python
#visualizing the data for long trade
plt.figure(figsize = (50,20))
plt.plot(data['Close Price'], label='data')
plt.plot(SMA['SMA30'], label = 'SMA30')
plt.plot(SMA['SMA300'], label = 'SMA300')
plt.title("SMA for LONG TRADE")
plt.xlabel('1980/01/02 - 2019/11/25')
plt.ylabel('closing price')
plt.legend()
plt.grid()
plt.show()
```

```python
#creating new dataframe 
df = pd.DataFrame()
df['Dates'] = data['Dates']
df['data'] = data['Close Price']                      #df['data'] is Closing Price of the stock
df['SMA30'] = SMA['SMA30']              
df['SMA300'] = SMA['SMA300']            
df['Dates'] = pd.to_datetime(df['Dates'])
df['year'] = df['Dates'].dt.year
```

```python
# MAIN STRATEGY which produces signal when to buy and sell

def buy_sell(df):
    profitable_long_trades = 0           
    loss_long_trades = 0
    profitable_short_trades = 0
    loss_short_trades = 0
    SigPriceBuy = []
    SigPriceSell = []
    SigDates = []
    flag0 = -1
    flag1 = -1
    buy_price =0
    sell_price =0
    returns = 100
    maxx = 0
    minn = 0
    returns_array = []
    LongTrades = []
    ShortTrades = []
    Annual_return = [100]*40
    Annual_return_array = [0]*40             #list to store Annual Returns for every year
    
    for i in range(len(df)):
        if (df['SMA30'][i] > df['SMA300'][i]) and (df['SMA30'][i-1] < df['SMA300'][i-1]):   # opening long position
                SigPriceBuy.append(df['data'][i])
                SigPriceSell.append(np.nan)
                SigDates.append(df['Dates'][i])
                LongTrades.append('buy_open')
                flag1 = 1
                maxx = df['data'][i]
                buy_price = df['data'][i]
                
                if flag0==0:                                    # closing the open short position due to signal reversal
                    ShortTrades.append('Closed_due_to_reversal_signal')
                else:
                    ShortTrades.append(np.nan)

        elif (df['SMA30'][i] < df['SMA300'][i]) and (df['SMA30'][i-1] > df['SMA300'][i-1]):     # opening short position
                SigPriceBuy.append(np.nan)
                SigPriceSell.append(df['data'][i])
                SigDates.append(df['Dates'][i])
                ShortTrades.append('short_open')
                flag1 = 0
                minn = df['data'][i]
                sell_price = df['data'][i]
                 
                if flag0==1:                                   # closing the open long position due to signal reversal
                    LongTrades.append('Closed_due_to_reversal_signal')
                else:
                    LongTrades.append(np.nan)
                
        if flag1==1 and flag0==1:                                             #checking stoploss each time if long position is opened
                SigPriceBuy.append(np.nan)
                SigPriceSell.append(np.nan)
                SigDates.append(np.nan)
                ShortTrades.append(np.nan)
                if maxx < df['data'][i]:
                    maxx = df['data'][i]
                if df['data'][i] <= 0.89*maxx:                  # closing the open long position if price hits the stoploss
                    sell_price = df['data'][i]
                    LongTrades.append('Closed_due_to_stoploss')
                    flag1=-1
                else:
                    LongTrades.append('opened')
        
        elif flag1==0 and flag0==0:                                           #checking stoploss each time if short position is opened
                SigPriceBuy.append(np.nan)
                SigPriceSell.append(np.nan)
                SigDates.append(np.nan)
                LongTrades.append(np.nan)
                if df['data'][i] >= 1.02*minn:                 # closing the open short position if price hits the stoploss
                    buy_price = df['data'][i]
                    ShortTrades.append('Closed_due_to_stoploss')
                    flag1 = -1
                else:
                    ShortTrades.append('opened')
                if minn > df['data'][i]:       
                    minn = df['data'][i]   
                
        
        elif flag1==-1 and flag0==-1:                                         # if all positions are closed
                SigPriceBuy.append(np.nan)
                SigPriceSell.append(np.nan)
                SigDates.append(np.nan)
                ShortTrades.append(np.nan)
                LongTrades.append(np.nan)
        
        
    # computing returns, no. of profitable and loss trades!!!
        if flag0==1 and flag1!=1:                               #computing returns, after closing the open long position
            returns += returns*((sell_price - buy_price)/buy_price)
            Annual_return[df['year'][i] - 1980] += Annual_return[df['year'][i] - 1980]*((sell_price - buy_price)/buy_price)
            Annual_return_array[df['year'][i] - 1980] = Annual_return[df['year'][i] - 1980] - 100
            if sell_price - buy_price>0:
                profitable_long_trades+=1
            else:
                loss_long_trades+=1
        
        elif flag0==0 and flag1!=0:                               #computing returns, after closing the open short position                                     #shorting trades
            #returns += returns*((sell_price-buy_price)/sell_price)
            Annual_return[df['year'][i] - 1980] += Annual_return[df['year'][i] - 1980]*((buy_price - sell_price)/sell_price)
            Annual_return_array[df['year'][i] - 1980] = Annual_return[df['year'][i] - 1980] - 100
            if sell_price - buy_price>0:
                profitable_short_trades+=1
            else:
                loss_short_trades+=1
                
                
        flag0 = flag1
        returns_array.append(returns-100)
    print('profitable_long_trades = ' , profitable_long_trades)
    print('loss_long_trades = ',loss_long_trades)
    print('Total long trades = ',profitable_long_trades+loss_long_trades)
    
    print('profitable_short_trades = ', profitable_short_trades)
    print('loss_short_trades = ',loss_short_trades)
    print('Total short trades = ',profitable_short_trades+loss_short_trades)
    
    print('algo returns = ', returns-100,'%')
    print('market returns = ', 100*(df['data'][len(df)-1] - df['data'][0])/df['data'][0],'%')
    print('\n\nWINRatio = ', (profitable_long_trades+profitable_short_trades)/(loss_long_trades+loss_short_trades+profitable_long_trades+profitable_short_trades))
    
    return (SigPriceBuy, SigPriceSell, SigDates, ShortTrades, LongTrades, returns_array, Annual_return_array)
            
```

```python
#store the buy and sell data into the variable
buy_sell = buy_sell(df)
df['Buy_Signal'] = buy_sell[0]
df['Sell_Signal'] = buy_sell[1]
df['Signal_On_Dates'] = buy_sell[2]
df['Updates_On_ShortTrades'] = buy_sell[3]                   # also contains the reason for closing the open short position
df['Updates_On_LongTrades'] = buy_sell[4]                    # also contains the reason for closing the open long position  
df['Cummulative_Algo_Returns'] = buy_sell[5]
Algo_Annual_returns = buy_sell[6]                            # percentage annual algo returns
```

```python
#Calculating Cumulative market returns
Market_returns = 100
Cummulative_Market_returns = [100]
for i in range(1,len(df)):
    Market_returns = 100*(df['data'][i] - df['data'][0])/df['data'][0]
    Cummulative_Market_returns.append(Market_returns)

df['Cummulative_Market_returns'] = Cummulative_Market_returns
```

```python
#visualize the buy signals and the strategy for long trade
plt.figure(figsize=(50,20))
plt.scatter(df.index, df['Buy_Signal'], label='buy', marker='^', s=150, c='black')
plt.scatter(df.index, df['Sell_Signal'], label='sell', marker='o', s=150, c='green')
plt.plot(data['Close Price'], label='data', alpha=0.7, c='cyan')
plt.plot(SMA['SMA30'], label = 'SMA30', alpha=0.7)
plt.plot(SMA['SMA300'], label = 'SMA300', alpha=0.7, c='red')
plt.title("SIGNALS for TRADES")
plt.xlabel('1980/01/02 - 2019/11/25')
plt.ylabel('closing price')
plt.legend()
plt.grid()
plt.show()
```

```python
#visualising_returns
plt.plot(df['Cummulative_Market_returns'], label = 'market returns')
plt.plot(df['Cummulative_Algo_Returns'], label = 'algo returns')
plt.legend()
plt.grid()
plt.show()
```

```python
#Calculating percentage Annual Returns
Market_Annual_returns = []          
year = [1980]
prev_year_index = 0
for i in range(1,len(df)):
    if df['year'][i-1]<df['year'][i]:
        year.append(df['year'][i])
        Market_Annual_returns.append( 100*(df['data'][i-1] - df['data'][prev_year_index])/df['data'][prev_year_index] )
        prev_year_index = i
                
Market_Annual_returns.append( 100*(df['data'][10063] - df['data'][prev_year_index])/df['data'][prev_year_index] )

#VISUALISING ANNUAL RETURNS
plt.figure(figsize=(15,5))
plt.scatter(year, Market_Annual_returns, label='Market_Annual_returns', marker='^', s=100, c='black')
plt.scatter(year, Algo_Annual_returns, label='Algo_Annual_returns', marker='o', s=100, c='green')
plt.plot(year, Algo_Annual_returns, label='Algo_Annual_returns', alpha = 0.3)
plt.plot(year, Market_Annual_returns, label='Market_Annual_returns', alpha = 0.3)
plt.xlabel('year')
plt.ylabel('Annual_returns')
plt.legend()
plt.grid()
plt.show()
```

```python
#Now Calculating Annualised Returns
Annualised_Market_returns = (1+df['Cummulative_Market_returns'][10063]/100)**(365/10064) - 1
Annualised_Algo_returns = (1+df['Cummulative_Algo_Returns'][10063]/100)**(365/10064) - 1

print('Annualised_Market_returns = ',Annualised_Market_returns*100)
print('Annualised_Algo_returns = ',Annualised_Algo_returns*100)
```

```python
#Calculating Annualised Volatility
#Since we already have percentage annual returns
# Annualised Volatility = STD of percentage annual returns
Annualised_Volatility = np.std(np.array(Market_Annual_returns))
print("Annualised_Volatility_of_the_Market = ", Annualised_Volatility)
```

```python
#Calculating Rolling Returns
df['month'] = df['Dates'].dt.month
monthly_returns = []
twelwe_months_rolling_returns_of_the_market = pd.DataFrame()
prev_month_index = 0

for i in range(1,len(df)):
    if df['month'][i-1]<df['month'][i]:
        monthly_returns.append( 100*(df['data'][i-1] - df['data'][prev_month_index])/df['data'][prev_month_index] )
        prev_month_index = i

monthly_returns.append( 100*(df['data'][10063] - df['data'][prev_month_index])/df['data'][prev_month_index] )

monthly_return = pd.DataFrame()
monthly_return['monthly_returns'] = monthly_returns

twelwe_months_rolling_returns_of_the_market['rolling returns'] = monthly_return['monthly_returns'].rolling(window=12).mean()
```

```python
plt.figure(figsize=(15,5))
plt.plot(twelwe_months_rolling_returns_of_the_market.index, twelwe_months_rolling_returns_of_the_market['rolling returns'], label = 'rolling Returns')
plt.xlabel('index')
plt.ylabel('Market_Rolling_Returns')
plt.grid()
plt.show()
```
